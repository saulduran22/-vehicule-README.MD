```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Parameters
alpha = 1.5
beta = 0.7
delta = 0.2
gamma = 0.7

params = [alpha, beta, delta, gamma]

# Initial conditions
y0 = [30, 2]

# Time span
t_span = np.linspace(0, 40, 500)  # More points for better representation

# System of ordinary differential equations (ODEs)
def myODE(y, t, params):
    alpha, beta, delta, gamma = params
    X, Y = y
    
    dX = alpha * X - beta * X * Y
    dY = delta * X * Y - gamma * Y
    
    return [dX, dY]

# Solve the ODEs
sol = odeint(myODE, y0, t_span, args=(params,))

# Plot the results
plt.subplot(2, 1, 1)
plt.plot(t_span, sol[:, 0], color="b")
plt.xlabel('Time')
plt.ylabel('Sheeps')

plt.subplot(2, 1, 2)
plt.plot(t_span, sol[:, 1], color="g")
plt.xlabel('Time')
plt.ylabel('Wolfs')

plt.tight_layout()
plt.show()

```
